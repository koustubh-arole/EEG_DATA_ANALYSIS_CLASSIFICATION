# LDA

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from scipy.stats import skew
from scipy.signal import welch
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
from collections import Counter

# Load dataset
file_path = "C:/Users/HP/Desktop/eeg related data/eeg_data.csv"
df = pd.read_csv(file_path)

# Identify feature columns (All except the last column)
eeg_features = df.columns[:-1]  # EEG feature columns
target_label = df.columns[-1]   # Target label ('epileptic_')

# Normalize EEG data (Min-Max Scaling)
df[eeg_features] = (df[eeg_features] - df[eeg_features].min()) / (df[eeg_features].max() - df[eeg_features].min())

# Check class distribution
print("Class distribution:", Counter(df[target_label]))

# Ensure at least two classes exist
if df[target_label].nunique() < 2:
    print("âŒ Not enough classes for classification. Check dataset balance.")
else:
    # Feature Extraction: Statistical & Spectral Features
    def extract_features(data):
        features = {}
        features['Mean'] = np.mean(data)
        features['Variance'] = np.var(data)
        features['Skewness'] = skew(data)
        f, psd = welch(data, fs=256)  # Assuming 256 Hz sampling rate
        features['PSD'] = np.mean(psd)
        return features

    # Apply feature extraction
    feature_data = df[eeg_features].apply(lambda x: extract_features(x), axis=1, result_type="expand")

    # Add target labels
    feature_data[target_label] = df[target_label]

    # Separate features and labels
    X = feature_data.drop(columns=[target_label])
    y = feature_data[target_label]

    # Determine LDA components dynamically
    n_classes = y.nunique()
    n_features = X.shape[1]
    lda_components = min(n_features, n_classes - 1)  # Ensure valid LDA components

    if lda_components > 0:
        print(f"Applying LDA with {lda_components} component(s)...")
        lda = LinearDiscriminantAnalysis(n_components=lda_components)
        X_reduced = lda.fit_transform(X, y)
        method = "LDA"
    else:
        print("LDA not possible (only 1 class detected). Using PCA instead.")
        pca = PCA(n_components=min(2, n_features))  # PCA with max 2 components
        X_reduced = pca.fit_transform(X)
        method = "PCA"

    # Convert to DataFrame dynamically based on components
    df_reduced = pd.DataFrame(X_reduced, columns=[f'{method}{i+1}' for i in range(X_reduced.shape[1])])
    df_reduced[target_label] = y

    # Train-Test Split
    X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size=0.2, random_state=42)

    # Train Random Forest Classifier
    clf = RandomForestClassifier(n_estimators=100, random_state=42)
    clf.fit(X_train, y_train)

    # Predictions & Evaluation
    y_pred = clf.predict(X_test)
    print(f"\nðŸ“Š Classification Results using {method}:")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))

    # Visualization of Reduced Data (Only if 2D)
    if X_reduced.shape[1] == 2:
        plt.figure(figsize=(8,6))
        for label in df_reduced[target_label].unique():
            subset = df_reduced[df_reduced[target_label] == label]
            plt.scatter(subset[f'{method}1'], subset[f'{method}2'], label=f'Class {int(label)}')

        plt.xlabel(f'{method}1')
        plt.ylabel(f'{method}2')
        plt.title(f'{method} Projection of EEG Data')
        plt.legend()
        plt.show()
    else:
        print(f"Skipping visualization: {method} produced only {X_reduced.shape[1]} component(s).")
```

# DEFINING PARAMETERS

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'n_estimators': [50, 100, 200], 
    'max_depth': [None, 10, 20], 
    'min_samples_split': [2, 5, 10], 
    'min_samples_leaf': [1, 2, 4]
}

# Perform GridSearchCV
grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=3, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

# Print the best parameters
print("Best Parameters:", grid_search.best_params_)

# Use the best model
best_rf = grid_search.best_estimator_

# Evaluate the optimized model
y_pred_best = best_rf.predict(X_test)
print("Optimized Model Accuracy:", accuracy_score(y_test, y_pred_best))
print(classification_report(y_test, y_pred_best))
```
